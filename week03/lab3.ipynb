```python
import pandas as pd
city_data = pd.read_csv("city_data.csv")
ride_data = pd.read_csv("ride_data.csv")
city_data.head()
```

```python
join_data = city_data.merge(ride_data, left_on="city", right_on="city")
join_data.head()
```

```python
unique_cities = join_data["city"].unique()
driver_counts = []
ride_counts = []
average_fares = []

for city in unique_cities:
    driver_count = join_data.loc[join_data["city"] == city, "driver_count"].count()
    driver_counts.append(driver_count)
    ride_count = join_data.loc[join_data["city"] == city, "ride_id"].count()
    ride_counts.append(ride_count)
    average_fare = join_data.loc[join_data["city"] == city, "fare"].mean()
    average_fares.append(average_fare)

interpreted_data = pd.DataFrame(data={"city": unique_cities, "average_fare": average_fares, "driver_count": driver_counts, "ride_count": ride_counts}).sort_values("city")
```

```python
# Plot with bubble chart
interpreted_data.plot.scatter(x="ride_count", y="average_fare", s=interpreted_data["ride_count"] * 10, c = 'green', alpha = 0.5).set_title("Rides vs. Fares")
```

```python
unique_types = join_data["type"].unique()
total_fares = sum(join_data["fare"])
total_rides = len(join_data)
total_by_type = []
total_by_rides = []

for unique_type in unique_types:
    total_type = join_data.loc[join_data["type"] == unique_type, "fare"].sum()
    total_rides = len(join_data.loc[join_data["type"] == unique_type, "ride_id"])
    total_by_type.append(total_type)
    total_by_rides.append(total_rides)

totals_fares = pd.DataFrame(data={"total": total_by_type}, index=unique_types)
pie_fares = totals_fares.plot(kind="pie", subplots=True, autopct='%1.1f%%')

totals_rides = pd.DataFrame(data={"total": total_by_rides}, index=unique_types)
pie_rides = totals_rides.plot(kind="pie", subplots=True, autopct='%1.1f%%')
```

```python
# Dimensionality Reduction - PCA on Iris
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.decomposition import PCA as sklearnPCA
import sklearn.preprocessing as preprocessing
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
data = pd.read_csv(url, header=None)

y = data[4]
x = data.iloc[:, 0:4]
```

```python
plt.scatter(x[y == 'Iris-setosa'].iloc[:, 0], x[y == 'Iris-setosa'].iloc[:, 1], label='Iris-setosa', c='red')
plt.scatter(x[y == 'Iris-versicolor'].iloc[:, 0], x[y == 'Iris-versicolor'].iloc[:, 1], label='Iris-versicolor', c='blue')
plt.scatter(x[y == 'Iris-virginica'].iloc[:, 0], x[y == 'Iris-virginica'].iloc[:, 1], label='Iris-virginica', c='lightgreen')

plt.legend()
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')

plt.show()
```

```python
x_norm = (x - x.min())/(x.max() - x.min())
```

```python
plt.scatter(x_norm[y == 'Iris-setosa'].iloc[:, 0], x_norm[y == 'Iris-setosa'].iloc[:, 1], label='Iris-setosa', c='red')
plt.scatter(x_norm[y == 'Iris-versicolor'].iloc[:, 0], x_norm[y == 'Iris-versicolor'].iloc[:, 1], label='Iris-versicolor', c='blue')
plt.scatter(x_norm[y == 'Iris-virginica'].iloc[:, 0], x_norm[y == 'Iris-virginica'].iloc[:, 1], label='Iris-virginica', c='lightgreen')

plt.legend()
plt.xlabel('Feature A')
plt.ylabel('Feature B')

plt.show()
```

```python
pca = sklearnPCA(n_components = 2)
transformed = pd.DataFrame(pca.fit_transform(x_norm))
```

```python
plt.scatter(transformed[y == 'Iris-setosa'].iloc[:, 0], transformed[y == 'Iris-setosa'].iloc[:, 1], label='Iris-setosa', c='red')
plt.scatter(transformed[y == 'Iris-versicolor'].iloc[:, 0], transformed[y == 'Iris-versicolor'].iloc[:, 1], label='Iris-versicolor', c='blue')
plt.scatter(transformed[y == 'Iris-virginica'].iloc[:, 0], transformed[y == 'Iris-virginica'].iloc[:, 1], label='Iris-virginica', c='lightgreen')

plt.legend()
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')

plt.show()
```

```python
# Exercises

wine_data = pd.read_csv("wine.data.csv")
label = wine_data["Label"]
wine_data = wine_data.drop(columns = "Label")
```

```python
sns.pairplot(data = wine_data)
```

```python
wine_corr = wine_data.corr()

plt.figure(figsize = (12,5))

sns.heatmap(wine_corr, xticklabels = wine_corr.columns, yticklabels = wine_corr.columns, annot=True)

# High correlation pairs:
# - OD280 and Flavanoids
# - Flavanoids and Total phenols
```

```python
standard_scaler = preprocessing.StandardScaler()
standard_scaler.fit(wine_data)
x_scaled_array = standard_scaler.transform(wine_data)
normalized_data = pd.DataFrame(x_scaled_array, columns = wine_data.columns)
```

```python
from sklearn.cluster import KMeans

k_means_clustering = KMeans(n_clusters = 3, random_state = 250899)
clusters = k_means_clustering.fit_predict(normalized_data)

normalized_data["cluster"] = clusters.astype('float64')
sns_plot = sns.pairplot(normalized_data, hue="cluster", diag_kind="hist")
```

```python
from sklearn.decomposition import PCA as sklearnPCA
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
pca_count = 6

pca = sklearnPCA(n_components=pca_count)
transformed_data = pd.DataFrame(pca.fit_transform(normalized_data))
clusters_transformed = k_means_clustering.fit_predict(transformed_data)

original_kmeans = adjusted_rand_score(label, clusters)
pca_kmeans = adjusted_rand_score(label, clusters_transformed)

print("Original kMeans score: " + str(original_kmeans))
print("Transformed kMeans score: " + str(pca_kmeans))
# This rand index is close to a perfect match of 1.0 so we can suggest that the k-means
#  method of identifying clusters is effective.
```
